## Define the Following:

### Input Layer: 
The entry points for the data into the neural network. These are the nerons that first touch and process the input data.
### Hidden Layer:
These are neurons that take the output of a previous neurons and use that output as they're input. These neurons never handle or process the data directly.
### Output Layer:
These are nuerons that process the work done by the other layers and give a meaningful, interpretable result.

### Neuron:
At it's heart, little more than a mathematical function. It takes an input, runs it through a certain number of calculations, and gives an output.
### Weight:
Once a layer has finished its process and passes the data onto the next layer, weights tell the next layer how important each bit of processed data is.
### Activation Function:
Some neurons aren't always run every time a neural network processes data. Conditions under which certain neurons which would normally be unrun will actually be run are activation functions.
### Node Map:
The overall architecture of a neural network. How many layers there are, how many nodes per layer, etc. 
### Perceptron:
The simplest kind of neural network. It consits of one input node that takes as many inputs as you desire, runs some computations on those inputs, and then will return an output.


## Inputs -> Outputs

### Explain the flow of information through a neural network from inputs to outputs. Be sure to include: inputs, weights, bias, and activation functions. How does it all flow from beginning to end?

Data enters the input layer nodes and the input layer nodes process the data. It does this by running all nodes whose activation functions are met. Once all activated nodes are run the data is sent to the next layer. On the way, as the data goes from the input layer to the first hidden layer two things happen: the weight of the connection is applied and bias is added (IN THAT ORDER). This process repeats for all remaining non-output layers. 1) run all activated nodes whose activation functions are satisfied, 2) adjust the output of these activated nodes by first the weight of the connection and then the bias of the layer, and 3) send the finished process to the next layer. Once that is done for all non-output layer layers, the output layer takes the data it receives and processses it, after which the results will be an output that can be applied to the data i.e the output is meaningful in the context of the problem, there is no further processing that needs to be done to reveal the meaning of the output.

```python
import pandas as pd
data = { 'x1': [0,1,0,1],
         'x2': [0,0,1,1],
         'bias': [1,1,1,1],
         'y':  [1,1,1,0]
       }

df = pd.DataFrame.from_dict(data).astype('int')
df.head()
```

```python
import numpy as np

def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1 - sx)
    
def sigmoid_derivative2(x):
    return (np.exp(-x) * sigmoid(x)**2)
```

```python
# get our inputs and output from the dataframe
inputs = df.to_numpy()

output_badshape = df.iloc[:4, -1].to_numpy()
correct_output = []
for i in output_badshape:
    correct_output.append([i])
correct_output = np.array(correct_output)


print(inputs, "\n\n", correct_output)
```

```python
# make weights
from numpy import random

# gets a list of numbers from -1 to 1
# we have 2 input vectors, so we need 2 weights
weights = 2 * random.random_sample((4,1)) - 1
weights
```

```python
# apply weights
# we want a 4x1 at the end of this dot product. 
# inputs a 4x2, weights is a 2x1 so the desired (and only valid) order for the dot product is:
# inputs x weights
weighted_data = np.dot(inputs, weights)
weighted_data
```

```python
# act on weighted data by sigmoid function
activated_data = sigmoid(weighted_data)
activated_data
```

```python
error = correct_output - activated_data
error
```

```python
adjusted_error = error * sigmoid_derivative(activated_data)
adjusted_error
```

```python
# inputs 4x2, adjadjusted_error 4x1, 
# inputs.T 2x4
# valid product: inputs.T dot adjusted_error

weights += np.dot(inputs.T, adjusted_error)
weights
```

```python
### every step in one, done a bunch of times

# make random weights from -1 to 1
weights = 2 * random.random_sample((4,1)) - 1

for i in range(10000):
    # apply weights to input data
    weighted_data = np.dot(inputs, weights)
    # apply node to weighted data
    activated_data = sigmoid(weighted_data)
    # calculate error
    error = correct_output - activated_data
    # calculate necessary adjustments to the weights
    weight_adjustment = error * sigmoid_derivative(activated_data)
    # apply weight adjustment to weights
    weights += np.dot(inputs.T, weight_adjustment)

print("Weights: \n", weights)
print("\n Output: \n", activated_data)
print("\n Desired Output: \n", correct_output)
print("\n Error: \n", correct_output - activated_data)
```
